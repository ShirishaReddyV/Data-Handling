```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={ "Date" : pd.to_datetime(["2022-01-01","2022-02-01","2003-07-02","2004-08-05","2004-06-23","2004-01-20","2004-09-25"]),"value":[10,19,100,97,92,95,91]}
```

```python
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='*')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series With Trend")
plt.grid(True)
plt.show()
```
