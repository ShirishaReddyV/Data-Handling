```python
import requests
```

```python
Link="https://www.mercedes-benz.co.in/passengercars.html?group=all&subgroup=all.saloon&view=BODYTYPE&gagcmid=GA_20685282646_153411963263_677718176048&gad_source=1&gclid=CjwKCAjw48-vBhBbEiwAzqrZVHoCvCELzer2gm8ynFbuNzqYCkNBo6gPml4JUxBNx6ON4F27Vc-f5RoCVRcQAvD_BwE&gclsrc=aw.ds"
Link_text= requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
Soup = BeautifulSoup(Link_text, 'html.parser')
images = Soup.find_all('img')
print(images)
for item in images:
    print(item['src'])
```
